```python
# Use this cell to begin your analysis, and add as many as you would like!
# Importing requests, BeautifulSoup, nltk, and Counter
import requests
from bs4 import BeautifulSoup
import nltk
from collections import Counter
```

```python
# Getting the html file
book = requests.get("https://www.gutenberg.org/files/16/16-h/16-h.htm")
```

```python
# Setting the correct text encoding.
book.encoding = 'utf-8'
```

```python
# Extracting html
html_book = book.text
```

```python
print(html_book[:1500])
```

```python
# Getting text from html
# Creating BeautifulSoup object from html
soup_book = BeautifulSoup(html_book,'html.parser')
```

```python
# Getting the text out of the soup
text_book = soup_book.get_text()
```

```python
# Printing characters
print(text_book[:1500])
```

```python
# Creating a tokenizer
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
```

```python
# Tokenizing the text
tokens = tokenizer.tokenize(text_book)
```

```python
print(tokens[:20])
```

```python
# Makeing words lowers case
words = []
for word in tokens:
    words.append(word.lower())

print(words[:20])
```

```python
# Getting the English stop words from nltk.

sw = nltk.corpus.stopwords.words('english')

# Printing out the first twenty stop words.
print(sw[:20])
print(sw)
```

```python
# Removing stop words.
words_ns = []

for word in words:
    if word not in sw:
        words_ns.append(word)
print(words_ns[:20])
```

```python
# Initialize a Counter object from our processed list of words.
count = Counter(words_ns)

top_ten = count.most_common(10)

print(top_ten)
```
