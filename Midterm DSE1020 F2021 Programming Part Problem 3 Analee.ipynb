```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series, read_csv
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
mpl.rcParams['figure.figsize'] = (10, 6)
import seaborn as sns
import skimage
from skimage import data
```

```python
# 1: Load the coins data into a numpy array (data.coins())
coins = data.coins()
```

```python
# 2: display it using matplotlib as an image
plt.imshow(coins)
plt.colorbar()
```

```python
coins.shape
```

```python
# 3: Crop the coin data so that the top 12 coins (first two rows are visible)
coins_cropped = coins[0:150, :]
plt.imshow(coins_cropped)
```

```python
# 4: Use the ">" or "<" to create a numpy array which shows the values above a size you determine (threshold).
# Find a threshold value so that as much of the coins are "1" ie true, as possible while the background is still dark.
# You may want to use the histogram to pick that value (between 0 and 255).

plt.hist(coins_cropped.ravel(), bins=256, color='gray')
plt.show()

thresh = 130
binary_coins = coins_cropped > thresh
plt.imshow(binary_coins)
```

```python
# 5: Use the sum opperator to count the number of pixels
num_pix = np.sum(coins_cropped >= 0)
num_pix
```

```python
# 6: Use the number of coins you can see (12)
# and the formula for the area of the circle to calculate the average radius.
# Compare that with what you see on the screen.
import math

dark_space = np.sum(binary_coins == False)
coins_space = np.sum(binary_coins == True)
num_coins = 12

avg_coin_area = coins_space/num_coins
avg_radius = np.sqrt(avg_coin_area/math.pi)

print(avg_radius)
```

The above gives an average radius of ~21 pixels per coin. Based on the image itself, this looks like it's pretty close to what the actual radius is.
