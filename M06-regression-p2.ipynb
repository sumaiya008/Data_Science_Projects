```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
mpl.rcParams['figure.figsize'] = (9,5)
mpl.rcParams['figure.dpi']=100
```

```python
num_points = 100
np.random.seed(42)
xs = np.random.random((num_points,))
xs
```

```python
noise_level = 0.3
es = np.random.randn(num_points)*noise_level
m = 2.71828
b = 3.14159
ys = m*xs + b + es # observed measurements
```

```python
print(xs.shape, len(xs), xs.min(), xs.max())
```

```python
print(ys.shape, len(ys), ys.min(), ys.max())
```

```python
print(es.shape, len(ys), es.min(), es.max())
```

```python
plt.scatter(xs,ys)
plt.plot([0,1],[b,m+b],'b-')
```

```python
X = np.vstack([xs, np.ones(len(xs))]).T
X.shape
```

```python
X[:10,:]
```

```python
ys.shape
```

```python
Y = ys.reshape(len(ys),1)
Y.shape
```

```python
Y[0:3]
```

$$ X^T X C = X^T Y \implies C = (X^T X)^{-1} X^T Y $$

```python
Chat = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T, Y))
```

```python
Chat
```

```python
print('Computed: m={}, b={},  actual m={}, b={}'.format(Chat[0][0],Chat[1][0], m, b))
```
