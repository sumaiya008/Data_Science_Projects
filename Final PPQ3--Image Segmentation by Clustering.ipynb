Lets do some **unsupervised** learning. We are going to segment an image into regions via kmeans clustering. 

url = "https://bit.ly/coloredpowders"


## Import the image 

Use scipy.ndimage.imread to get a 275 Ã— 183 (rows and columns) x 3 (colors image)


```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
from imageio import imread
image = imread('https://bit.ly/coloredpowders')
plt.imshow(image)
```


## Reformat image

Reshape the data so it has 275*183 rows and 3 columns. We are used to columns as being features and rows (pixels) as being different examples. This does that.


```python
df=pd.DataFrame(image.reshape(-1, 3),columns=['r', 'g', 'b'])
df.head()
```

```python
df.shape
```

## EDA with the first 2 columns

We can't look at all the data but lets look at some


```python
import seaborn as sns
plt.figure(figsize=(8,8))
plt.subplot(221)
sns.histplot(df.r , color ='purple', kde=True)

plt.subplot(222)
sns.histplot(df.b , color ='red', kde=True)
```

## Find three region map

Run K-means clustering on the data with k=3 to get three clusters


```python
# Import KMeans
from sklearn.cluster import KMeans
 
# Create a KMeans instance with 3 clusters: model
model = KMeans(n_clusters=3)
 
# Fit model to points
model.fit(df)
 
# Determine the cluster labels of new_points: labels
labels = model.predict(df)
 
# Print cluster labels of new_points
print(labels)
```
