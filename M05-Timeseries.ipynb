```python
bike_data = pd.read_csv('FremontBridge.csv', index_col='Date', parse_dates=True)
bike_data.head()
```

```python
bike_data.info()
```

```python
bike_data.plot(alpha=0.2)
plt.ylabel('Hourly Bike Count')
```

```python
weekly = bike_data.resample('W').sum()
weekly.plot(style= [':','--','-'])
plt.ylabel('Weekly bicle count')
```

```python
daily = bike_data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style= [':','--','-'])
plt.ylabel('mean hourly count')
```

```python
daily = bike_data.resample('D').sum()
daily.rolling(50, center=True, win_type='gaussian').sum(std=10).plot(style= [':','--','-'])
plt.ylabel('mean hourly count')
```

```python
by_time = bike_data.groupby(bike_data.index.time).mean()
hourly_ticks = 4*60*60*np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':','--','-'])
```
