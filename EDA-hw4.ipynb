Now i want to understand from dataset that either girls are doing good in performance or boys. For that first let's make a subset Final grade then we will compare it with other variables.

### Final Grade

```python
st_data['Final_G'] = (st_data["G1"]+st_data["G2"]+st_data["G3"])/3
```

```python
st_data['Final_G'].head()
```

### Study time Vs Grades

```python
sns.barplot(x="studytime",y="Final_G", hue="sex", data=st_data, ci=False)
```

Male students are studying more and doing better on tests.

```python
sns.scatterplot(st_data, x="absences",y='Final_G',hue='sex')
```

Studens who have less absences are better in grade performance.

### Failures Vs Grade

```python
sns.barplot(x="failures", y="Final_G", hue="sex", data=st_data, ci=False)
```

Male students have less failures.

### School Vs Grade

```python
sns.barplot(x="school", y="Final_G", hue="sex", data=st_data, ci=False)
```

GP school have more better grades and male students are on high point.

### Travel time Vs Grade

```python
sns.barplot(x="traveltime", y="Final_G", hue="sex", data=st_data, ci=False)
```

Students who lives nearby school are doing good. In this case also Male students have better grades.

## Summery

From overall observation statistically we can say Male Students are scored better than Female Students.
