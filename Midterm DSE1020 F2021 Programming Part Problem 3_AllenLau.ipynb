```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series, read_csv
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
mpl.rcParams['figure.figsize'] = (10, 6)
import seaborn as sns
import skimage
```

```python
#Load Coins Data in numpy array 
coins_np = skimage.data.coins()
```

```python
#Display coin image using matplotlib
plt.imshow(coins_np)
```

```python
#crop coins using array slicing 
coins_crop = coins_np[:150]

#show image
plt.imshow(coins_crop)
```

```python
#threholding 

thresh = 140

coins_thres = coins_crop > thresh
plt.imshow(coins_thres)
```

```python
#Pixel Count 

coins_thres.sum()
```

```python
plt.hist(coins_thres)
```

```python
#Logic for finding area written out 
#find the number of pixels that are coins based on the intensity value of the numpy array
#calculate the radius based this value and the fact that we have 12 coins. 

avg_area_coin = coins_thres.sum()/12 
r = np.sqrt(avg_area_coin/np.pi)
```

```python
#average radius of coin in pixels
r
```
